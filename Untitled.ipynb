#**Data Cleaning**


```python
# Gathered Information
# 1. Hompage has too much null information and has to be dropped
# 2. Columns like tagline, keywords, production_companies, drectors will have to be dropped
# 3. The data has one duplicated item and will be dropped. 
```

```python
df_movies.shape
```

```python
df_movies.drop_duplicates(inplace = True)
```

```python
# Checking if the duplicate has been dropped
df_movies.shape
```

```python
DroppedColumns = ['homepage', 'tagline', 'overview' ,'cast','director',
       'keywords','overview','production_companies','release_date']
```

```python
df_movies.drop(DroppedColumns, axis=1, inplace=True)
```

```python
df_movies.head()
```

```python
#Checking for Null Value
df_movies.isnull().sum()
```

```python
# Checking the genre in which it is null
df_movies[df_movies.genres.isnull()]
```

```python
df_movies.dropna(inplace = True)
```

# 1  Which movie title has the highest budget ?

```python
Top5Movies= df_movies[['original_title', 'budget']].sort_values(by='budget', ascending=False).head()
```

```python
Top5Movies
```

```python
Top5Movies.set_index('original_title').plot(kind ='barh')
plt.title('Top 5 Movies Ranked by Budget');
```

# 2 Which movie titles has the highest revenue?

```python
Top5MoviesRevenue= df_movies[['original_title', 'revenue']].sort_values(by='revenue', ascending=False).head()
```

```python
Top5MoviesRevenue
```

```python
Top5MoviesRevenue.set_index('original_title').plot(kind ='barh')
plt.title('Top 5 Movies Ranked by Revenue');
```
